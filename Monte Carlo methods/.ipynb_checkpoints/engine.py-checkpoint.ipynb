```python
from importlib import reload
import car; reload(car)
import environment; reload(environment)
Environment = environment.Environment
Car = car.Car
```

```python
import random
def select_action_fn(car):
    action = None
    if car.P.get((car.position, car.speed), None): # and random.random() <= car.greedy_probability:
        action = car.P[(car.position, car.speed)]

    while True:
        if action is None or car.speed[0] + action[0] > 0 or car.speed[1] + action[1] < 0:
            action = random.choice(car.actions)
        else:
            break

    if random.random() > 0.1:
        return action
    else:
        return (0, 0)

env = Environment("racetrack2_30x32.txt")
car = Car(env, select_action_fn=select_action_fn)
```

```python
car.P = {}
car.train(100, update_policy_each=20)
```

```python
car._reset()
path = []
path.append(car.position)
for i in range(1000):
    reward, action = car.step()
    path.append(car.position)
    if env.is_finish(car.position): break

print(len(path), path[:4])
env.print(car, path)
    
```
